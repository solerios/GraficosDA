```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Supongamos que tenemos un DataFrame de Pandas df que contiene datos de ventas por región y mes:

```python


# Ejemplo de DataFrame
data = {
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Mes': ['Enero', 'Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Febrero'],
    'Ventas': [100, 150, 120, 110, 130, 140, 160, 155]
}

```

Generamos el df y graficamos:

```python
df = pd.DataFrame(data)

# Creación de un gráfico de barras usando Seaborn
sns.barplot(x='Region', y='Ventas', data=df, hue='Mes')
plt.title('Ventas por Región y Mes')
plt.show()
```
