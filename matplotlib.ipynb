# **Matplotlib**

Es la base para aprender visualizacion de datos aunque haya bibliotecas más sofisticadas

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Conceptos basicos
### Arquitectura

```python
#Datos de ejemplo

x= np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure (figsize = (10,6))
#la priera linea determina el tamanio del grafico
ax = fig.add_subplot(1 , 1 , 1)
#Axes es el area del grafico, los axes se agregan a la figure(dodne se van a ;dibujar)
ax.set_xlim(0,10)
ax.set_ylim(-1.5, 1.5)
#Estios son los ejes controlan el rango y las etiquetas
line, = ax.plot(x, y, label = 'Seno', color = 'green')
text = ax.text(2, np.sin(2), 'Punto', fontsize = 12, color = 'red')#Aca van los elementos graficos
ax.legend()
plt.show()
```

### Estilos

```python
#Imprimimos una liste de estilos disponibles en matplotlib
print("Estilos disponibles:", plt.style.available)
plt.style.use('ggplot') #Elegimos un estilo de los  disponibles
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(x,y, label = 'Seno', color = "blue")
ax.plot(x, y2, label = 'Coseno', color = 'red')
ax.set_title('grafico de Seno y coseno con estilo seaborn -darkgrid')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
plt.show()
```

```python
#Imprimimos una liste de estilos disponibles en matplotlib
print("Estilos disponibles:", plt.style.available)
plt.style.use('ggplot') #Elegimos un estilo de los  disponibles
fig, ax = plt.subplots(figsize = (10,6))
#ax = fig.add_subplot(1 , 1 , 1)
#Axes es el area del grafico, los axes se agregan a la figure(dodne se van a ;dibujar)
#ax.set_xlim(0,10)
ax.set_ylim(-1.0, 1.0)
ax.plot(x,y, label = 'Seno', color = "green")
ax.plot(x, y2, label = 'Coseno', color = 'orange')
ax.set_facecolor('lightgrey')
ax.grid(True, linestyle = '--', color = 'grey', alpha = 0.7)
ax.set_title('grafico de Seno y coseno con estilo seaborn -darkgrid')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
plt.show()
```
