### Gráficos 3D


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generar datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear figura y ejes 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar gráfico de superficie 3D
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
# plot_surface() dibuja la superficie 3D con datos X, Y, Z
# cmap='viridis' define el esquema de color (escala de colores)

# Añadir barra de color
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)  # Ajustes de la barra de color

# Configurar etiquetas y título
ax.set_title('Gráfico de Superficie 3D')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()  # Mostrar el gráfico 3D
```
