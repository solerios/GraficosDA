### Configuración de Escalas de Ejes
#### Escala Lineal
La escala lineal es la escala predeterminada en Matplotlib y se utiliza para datos que crecen de manera uniforme.

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y)
plt.yscale('linear')
plt.title('Escala Lineal')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

#### Escala Logarítmica
La escala logarítmica se utiliza para datos que crecen exponencialmente.

```python
plt.plot(x, y)
plt.yscale('log')
plt.title('Escala Logarítmica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

#### Escala Simétrica
La escala simétrica se utiliza para datos que tienen valores tanto positivos como negativos que necesitan ser tratados simétricamente.

```python
y = [-25, -16, -4, 0, 9, 25]

plt.plot(x, y)
plt.yscale('symlog')
plt.title('Escala Simétrica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

#### Escala Logit
La escala logit se utiliza principalmente en análisis de datos categóricos y probabilidades.

```python
y = [0.01, 0.1, 0.5, 0.9, 0.93, 0.99]

plt.plot(x, y)
plt.yscale('logit')
plt.title('Escala Logit')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

### Añadiendo y Personalizando Grillas
Añadir Grillas Básicas

```python
plt.plot(x, y)
plt.grid(True)
plt.title('Grilla Básica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

Personalización de las Grillas


```python
plt.plot(x, y)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Grilla Personalizada')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

Grillas Mayores y Menores


```python
plt.plot(x, y)
plt.grid(which='major', color='black', linestyle='-')
plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle=':', linewidth=0.5)
plt.title('Grillas Mayores y Menores')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

### Ejemplos Prácticos
Ejemplo Básico de Escalas de Ejes

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y)
plt.yscale('linear')
plt.title('Escala Lineal')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

Ejemplo Avanzado con Diferentes Escalas y Grillas Personalizadas


```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [-25, -16, -9, -4, 0, 4]  # Ajustado para que tenga longitud 6

# Definición de y_sym y y_logit
y_sym = [-16, -9, -4, 0, 16, 25]
y_logit = [0.01, 0.1, 0.5, 0.9, 0.93]  # Ajustado para que tenga longitud 5

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Escala Lineal con Grilla
axs[0, 0].plot(x, y)
axs[0, 0].set_yscale('linear')
axs[0, 0].grid(True)
axs[0, 0].set_title('Escala Lineal con Grilla')

# Escala Logarítmica con Grilla Personalizada
axs[0, 1].plot(x, y)
axs[0, 1].set_yscale('log')
axs[0, 1].grid(color='gray', linestyle='--', linewidth=0.5)
axs[0, 1].set_title('Escala Logarítmica con Grilla Personalizada')

# Escala Simétrica con Grillas Mayores y Menores
axs[1, 0].plot(x, y_sym)
axs[1, 0].set_yscale('symlog')
axs[1, 0].grid(which='major', color='black', linestyle='-')
axs[1, 0].minorticks_on()
axs[1, 0].grid(which='minor', color='gray', linestyle=':', linewidth=0.5)
axs[1, 0].set_title('Escala Simétrica con Grillas Mayores y Menores')

# Escala Logit con Grilla
axs[1, 1].plot(x[:5], y_logit)
axs[1, 1].set_yscale('logit')
axs[1, 1].grid(True)
axs[1, 1].set_title('Escala Logit con Grilla')

plt.tight_layout()
plt.show()
```
